## Cleaning Data
In this section, clean **all** of the issues you documented while assessing. 

**Note:** Make a copy of the original data before cleaning. Cleaning includes merging individual pieces of data according to the rules of [tidy data](https://cran.r-project.org/web/packages/tidyr/vignettes/tidy-data.html). The result should be a high-quality and tidy master pandas DataFrame (or DataFrames, if appropriate).

```python
# Make copies of original pieces of data
twitter_archive_clean = twitter_archive.copy()
image_predictions_clean = image_predictions.copy()
tweets_data_clean = tweets_data.copy()
```

```python
# -------------------------------------------
# Cleaning twitter_archive_clean
# -------------------------------------------

# 1. Remove retweets (rows with non-null 'retweeted_status_id')
twitter_archive_clean = twitter_archive_clean[twitter_archive_clean['retweeted_status_id'].isnull()]

# 2. Remove duplicates in 'tweet_id'
twitter_archive_clean = twitter_archive_clean.drop_duplicates(subset='tweet_id')

# 3. Convert 'timestamp' to datetime
twitter_archive_clean['timestamp'] = pd.to_datetime(twitter_archive_clean['timestamp'])

# 4. Clean dog names: Replace names that are not actual names (e.g., 'a', 'an', 'the') with None
invalid_names = ['a', 'an', 'the', 'None', 'none', 'O', 'old', 'very', 'quite', 'just', 'one', 'not', 'his', 'mad', 'such', 'unacceptable', 'infuriating', 'officially', 'light', 'my', 'all', 'space', 'by', 'this', 'getting', 'actually']
twitter_archive_clean['name'] = twitter_archive_clean['name'].replace(invalid_names, pd.NA)

# 5. Combine dog stages into a single column 'dog_stage'
def combine_stages(row):
    stages = []
    for stage in ['doggo', 'floofer', 'pupper', 'puppo']:
        if row[stage] == stage:
            stages.append(stage)
    return ','.join(stages) if stages else pd.NA

twitter_archive_clean['dog_stage'] = twitter_archive_clean.apply(combine_stages, axis=1)

# 6. Drop the original dog stage columns
twitter_archive_clean = twitter_archive_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)

# 7. Remove rows with missing or incorrect 'expanded_urls'
twitter_archive_clean = twitter_archive_clean[twitter_archive_clean['expanded_urls'].notnull()]

# 8. Ensure 'rating_numerator' and 'rating_denominator' are integers and positive
twitter_archive_clean = twitter_archive_clean[
    (twitter_archive_clean['rating_numerator'] > 0) &
    (twitter_archive_clean['rating_denominator'] > 0)
]

# -------------------------------------------
# Cleaning image_predictions_clean
# -------------------------------------------

# 1. Remove duplicates in 'tweet_id'
image_predictions_clean = image_predictions_clean.drop_duplicates(subset='tweet_id')

# 2. Remove rows where none of the predictions are dogs
image_predictions_clean = image_predictions_clean[
    image_predictions_clean[['p1_dog', 'p2_dog', 'p3_dog']].any(axis=1)
]

# 3. Standardize breed names: lowercase and replace underscores with spaces
for col in ['p1', 'p2', 'p3']:
    image_predictions_clean[col] = image_predictions_clean[col].str.lower().str.replace('_', ' ')

# 4. Remove rows with missing values (if any)
image_predictions_clean = image_predictions_clean.dropna()

# 5. Keep only images that correspond to tweets in twitter_archive_clean
image_predictions_clean = image_predictions_clean[
    image_predictions_clean['tweet_id'].isin(twitter_archive_clean['tweet_id'])
]

# -------------------------------------------
# Cleaning tweets_data_clean
# -------------------------------------------

# 1. Remove retweets (rows with non-null 'retweeted_status')
tweets_data_clean = tweets_data_clean[tweets_data_clean['retweeted_status'].isnull()]

# 2. Remove duplicates in 'id'
tweets_data_clean = tweets_data_clean.drop_duplicates(subset='id')

# 3. Convert 'created_at' to datetime if not already
if tweets_data_clean['created_at'].dtype != 'datetime64[ns, UTC]':
    tweets_data_clean['created_at'] = pd.to_datetime(tweets_data_clean['created_at'])

# 4. Keep only relevant columns for analysis
columns_to_keep = ['id', 'retweet_count', 'favorite_count', 'created_at']
tweets_data_clean = tweets_data_clean[columns_to_keep]

# 5. Remove rows with missing values in key columns
tweets_data_clean = tweets_data_clean.dropna(subset=['id', 'retweet_count', 'favorite_count', 'created_at'])

# 6. Ensure 'id' is of integer type
tweets_data_clean['id'] = tweets_data_clean['id'].astype('int64')

# -------------------------------------------
# Tidiness: Merge datasets into a master DataFrame
# -------------------------------------------

# Merge twitter_archive_clean and image_predictions_clean on 'tweet_id'
master_df = pd.merge(twitter_archive_clean, image_predictions_clean, how='inner', on='tweet_id')

# Merge with tweets_data_clean on 'tweet_id'/'id'
master_df = pd.merge(master_df, tweets_data_clean, how='inner', left_on='tweet_id', right_on='id')

# Drop duplicate 'id' column from tweets_data_clean
master_df = master_df.drop('id', axis=1)

# The master_df is now cleaned and tidy for analysis
```

## Analyzing and Visualizing Data
In this section, analyze and visualize your wrangled data. You must produce at least **three (3) insights and one (1) visualization.**

```python
import matplotlib.pyplot as plt

# Insight 1: Most common dog breed predicted (using p1)
top_breeds = master_df[master_df['p1_dog']].groupby('p1')['tweet_id'].count().sort_values(ascending=False).head(10)
print("Top 10 most common dog breeds (p1 prediction):")
print(top_breeds)

# Insight 2: Average rating numerator by dog stage
avg_rating_by_stage = master_df.groupby('dog_stage')['rating_numerator'].mean().dropna().sort_values(ascending=False)
print("\nAverage rating numerator by dog stage:")
print(avg_rating_by_stage)

# Insight 3: Correlation between retweet count and favorite count
correlation = master_df['retweet_count'].corr(master_df['favorite_count'])
print(f"\nCorrelation between retweet count and favorite count: {correlation:.2f}")

# Visualization: Top 5 most common dog breeds (bar plot)
plt.figure(figsize=(10,6))
top_breeds.head(5).plot(kind='bar', color='skyblue')
plt.title('Top 5 Most Common Dog Breeds (p1 Prediction)')
plt.xlabel('Dog Breed')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Explanation:
# The bar plot above shows the five most frequently predicted dog breeds in the dataset (using the first prediction p1).
# This helps us understand which breeds are most commonly featured or recognized by the image prediction model.
# The insights printed above provide further understanding of breed popularity, how dog stage relates to ratings,
# and the strong positive correlation between retweets and favorites, indicating that popular tweets tend to be both liked and shared.
```
